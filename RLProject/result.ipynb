```python
import json
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def load_result1():# 고정 환경 모델 사이즈별 성능 테스트
    file_names = [
    "size(0)",
    "size(10)",
    "size(100)",
    "size(1000)",
    "size(10000)",
    "size(100000)",
    ]
    results = []
    for file_name in file_names:
        with open(f'./data/고정 환경 - 통합 모델/{file_name}.json') as f:
            result = json.load(f)
        results.append(result)

    results = np.array(results).mean(1)
    df = pd.DataFrame(results)
    df["size"] = [file_name[5:-1] for file_name in file_names]
    return df

def load_result2():# 변동 환경 모델 사이즈별 성능 테스트
    file_names = [
    "size(0)",
    "size(10)",
    "size(100)",
    "size(1000)",
    "size(10000)",
    "size(100000)",
    ]
    results = []
    for file_name in file_names:
        with open(f'./data/변동 환경 - 통합 모델/{file_name}.json') as f:
            result = json.load(f)
        results.append(result)
    
    results = np.array(results).mean(1)
    df = pd.DataFrame(results)
    df["size"] = [file_name[5:-1] for file_name in file_names]
    return df
        
def load_result3(): # 버퍼 사이즈 별 성능

    file_names = [
        "size(1)",
        "size(3)",
        "size(5)",
        "size(7)",
        "size(10)",
        "size(20)",
        "size(30)",
        "size(50)",
        "size(70)",
    ]
    results = []
    for file_name in file_names:
        with open(f'./data/변동 환경 - 구분 모델/{file_name}.json') as f:
            result = json.load(f)
        results.append(result)

    results = np.array(results).mean(1) 
    df = pd.DataFrame(results)
    df["size"] = [file_name[5:-1] for file_name in file_names]
    return df

def load_result4(): # 환경 변화 감지 및 모델 초기화
    file_names = [
        "size(3, 10)",
        "size(3, 30)",
        "size(3, 50)",
        "size(3, 70)",
        "size(5, 10)",
        "size(5, 30)",
        "size(5, 50)",
        "size(5, 70)",
        "size(7, 10)",
        "size(7, 30)",
        "size(7, 50)",
        "size(7, 70)",
        "size(9, 10)",
        "size(9, 30)",
        "size(9, 50)",
        "size(9, 70)",
#         "a", "b", "c", "d"
    ]
    results = []
    for file_name in file_names:
        with open(f'./data/변동 환경 - 구분 모델 - 환경 감지/{file_name}.json') as f:
            result = json.load(f)
        results.append(result)

    results = np.array(results).mean(1) 
    df = pd.DataFrame(results)
    df["size"] = [file_name[5:-1] for file_name in file_names]
    return df
```

```python
df1 = load_result1()
df1["type"] = "고정환경 통합 모델"
df2 = load_result2()
df2["type"] = "변동환경 통합 모델"
df3 =  load_result3()
df3["type"] = "변동환경 구분 모델"
df4 =  load_result4()
df4["type"] = "변동환경 구분 모델 (환경 감지)"

df = pd.concat([df2, df3, df4])
```

```python
def moving_avg(arr, w):
    arr2 = []
    for i, x in enumerate(arr):
        f = int(max(0, i-w))
        t = int(min(x+i, len(arr)-1))
        arr2.append(np.array(arr[f:t+1]).mean())
    return arr2
```
